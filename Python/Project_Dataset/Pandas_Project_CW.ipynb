```python
data["Tamil"].fillna(T,inplace=True)
data["English"].fillna(E,inplace=True)
data["Maths"].fillna(M,inplace=True)
data["Science"].fillna(S,inplace=True)
data["Social Science"].fillna(SS,inplace=True)
data["Total Mark"].fillna(TM,inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data1=data.drop(["Student Name"],axis=1)
correlation_matrix = data1.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
X=data.drop(['Student Name',"Total Mark","Grade","Result"],axis=1)
Y=data["Total Mark"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```
