```python
X=data[['Age','Gender','Tot_Bilirubin','Direct_Bilirubin','Tot_Proteins','Albumin','Ag_Ratio','SGPT','SGOT','Alkphos']]
Y=data["Is_Patient"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=30)
```

```python
mpl.figure(figsize=(12,5))

#SGPT Histogram
mpl.subplot(1, 2, 1)
mpl.hist(data["SGPT"], bins=20, color="blue", alpha=0.7)
mpl.xlabel("SGPT")
mpl.ylabel("Count")
mpl.title("Distribution of SGPT")

#SGOT Histogram
mpl.subplot(1, 2, 2)
mpl.hist(data["SGOT"], bins=20, color="orange", alpha=0.7)
mpl.xlabel("SGOT")
mpl.ylabel("Count")
mpl.title("Distribution of SGOT")

mpl.tight_layout()
mpl.show()

# 2.Subplots for SGPT vs SGOT (scatter) + trendline in same figure

mpl.figure(figsize=(12,5))

# Left subplot → Scatter plot
mpl.subplot(1, 2, 1)
mpl.scatter(data["SGPT"], data["SGOT"], color="red", alpha=0.6, label="Patients")
mpl.xlabel("SGPT")
mpl.ylabel("SGOT")
mpl.title("SGPT vs SGOT (Scatter)")
mpl.legend()

# Right subplot → Scatter + Trendline
mpl.subplot(1, 2, 2)
mpl.scatter(data["SGPT"], data["SGOT"], color="red", alpha=0.6, label="Patients")

# Trendline
m, c = np.polyfit(data["SGPT"], data["SGOT"], 1)
mpl.plot(data["SGPT"], m*data["SGPT"] + c, color="blue", linestyle="--", label="Trendline")

mpl.xlabel("SGPT")
mpl.ylabel("SGOT")
mpl.title("SGPT vs SGOT with Trendline")
mpl.legend()

mpl.tight_layout()
mpl.show()
```
