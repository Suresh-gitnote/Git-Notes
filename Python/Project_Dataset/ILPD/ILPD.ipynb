```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

data["Gender"] = label_encoder.fit_transform(data["Gender"]) #Female = 0 && Male = 1
data
```

```python
data1=data.drop(["S.No."],axis=1)
correlation_matrix = data1.corr()

# Plot heatmap
mpl.figure(figsize=(12, 8))
sb.heatmap(correlation_matrix, annot=True, cmap='magma', linewidths=0.3)
mpl.title('Correlation Heatmap')
mpl.show()
```

```python
X=data[['Age','Gender','Tot_Bilirubin','Direct_Bilirubin','Tot_Proteins','Albumin','Ag_Ratio','SGPT','SGOT','Alkphos']]
Y=data["Is_Patient"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```
